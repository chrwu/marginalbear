```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
eval_result = pd.read_csv('eval0829.csv', delimiter='\s*,\s*', engine='python')
print(eval_result.shape)
eval_result.head()
```

```python
eval_result.mean(), eval_result.std()
```

```python
data = eval_result
data.head()
```

```python
# Create a figure instance
fig = plt.figure(1, figsize=(16, 12))

# Create an axes instance
ax = fig.add_subplot(111)
bp = ax.boxplot([data['random'], data['base'], data['pweight']],
                notch=False,
                vert=True,
                patch_artist=True
)
color = ['#7570b3', 'yellow', '#1b9e77']
for box, c in zip(bp['boxes'], color):
    box.set(facecolor=c, alpha=0.5, linewidth=3)

for med in bp['medians']:
    med.set(linewidth=3)
for cap in bp['caps']:
    cap.set(linewidth=3)
for whisker in bp['whiskers']:
    whisker.set(linewidth=3)
# for flier in bp['fliers']:
#     flier.set(marker='o', color='#e7298a', alpha=0.5)
ax.set_xticklabels(['Random', 'w/ Jaccard', 'w/ POS Ranking'])
ax.set_ylabel('nDCG', fontsize=30)
ax.set_ylim(-0.5, 0.5)
plt.title('Retrieval Evaluation with Similarity Ranking', fontsize=30)
ax.tick_params(labelsize=30)
# ax.set_ylim
```
